## Objective

Understand the gender distribution of students in the dataset

```python
import matplotlib.pyplot as plt 
import seaborn as sns
df["sex"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("gender")
plt.ylabel("Number of Students")
plt.show()
```

### Observation

- there are 383 females and 266 males 
- 59% are females and 41% are males 

### Insight

- The dataset contains more female students than male students
- The dataset is slightly imbalanced toward female students

### Possible Action

- No immediate action is required, as this graph only describes the gender distribution of the dataset

## Objective

Understand how students' final grades are distributed

```python
plt.figure(figsize=(8,5))
df["G3"].plot(kind="hist",bins=20)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.grid()
plt.show()
```

### Observation

- Most students scored between 10 and 15 marks
- Around 15 students failed (G3 = 0)
- Around 46 students scored 17 or above

### Insight

- Most students achieved average to above-average marks
- Very few students scored extremely low or extremely high marks
- The grade distribution is approximately bell-shaped, with most scores concentrated in the middle

### Possible Action

- Analyze the relationship between Study Time and G3.
- Analyze how previous failures affect the final grade.

## Objective

Analyze whether study time affects students final grades

```python
sns.barplot(data=df, x="studytime", y="G3")
plt.title("Study Time vs Final Grade")
plt.show()
```

### Observation

- Students with study time category 1 have an average final grade of 10.84
- Students with study time category 2 have an average final grade of 12.09
- Students with study time category 3 have an average final grade of 13.22
- Students with study time category 4 have an average final grade of 13.05

### Insight

- Average final grades generally increase as study time increases
- Students in study time categories 3 and 4 have similar average grades

### Possible Action

- Analyze how previous failures affect the final grade
- Investigate why students in study time category 4 do not have significantly higher grades than category 3

## Objective

Analyze the impact of previous failures on the final grade.

```python
sns.barplot(data=df,x="failures",y="G3")
plt.title("failures vs final grade")
plt.grid()
plt.show()
```

### Observation

- Students with 0 failures have an average final grade of 12.51
- Students with 1 failures have an average final grade of 8.64
- Students with 2 failures have an average final grade of 8.81
- Students with 3 failures have an average final grade of 8.07

### Insight

- Students with no previous failures have the highest average final grades
- Overall, final grades decrease as the number of previous failures increases
- Students with 1 and 2 failures have similar average final grades
- Students with 3 failures have the lowest average final grades

### Possible Action

- Investigate why students with 1 and 2 failures have similar average grades

## Objective

Identify relationships between numerical features

```python
# Correlation matrix
corr = df.corr(numeric_only=True)

# Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```

### Observation

- G3 has the strongest positive correlation with G2 (0.92) and G1 (0.83)
- Previous failures have a moderate negative correlation with G1, G2, and G3
- Study time has a weak positive correlation with students' grades

### Insight

- if G1 and G2 marks increases G3 also increases 
- if failures increases G3 decreases
- Previous academic performance (G1 and G2) is a strong predictor of the final grade (G3)

### Possible Action

- Provide additional guidance to students with previous failures
- Focus on factors that have stronger correlations with G3 rather than variables with weak correlations

## Objective

Detect potential outliers in numerical features

```python
columns = ["age", "studytime", "failures", "absences", "G1", "G2", "G3"]

plt.figure(figsize=(12,6))
sns.boxplot(data=df[columns])
plt.title("Box Plot of Numerical Features")
plt.xlabel("Features")
plt.ylabel("Values")

plt.grid()
plt.show()
```

### Observation

- Potential outliers are present in age, absences, G1, G2, and G3
- The (absences) column has the highest number of outliers
- Most features have a compact distribution, while absences shows high variability
- No impossible or invalid values are observed

### Insight

- The outliers appear to represent genuine student records rather than data entry errors
- High absence values may indicate students with irregular attendance
- Most of these outliers appear to be realistic and do not seem to be data entry errors

### Possible Action

- Retain the outliers because they represent genuine student records
- Recheck the effect of outliers after building machine learning models
- No outliers were removed at this stage
