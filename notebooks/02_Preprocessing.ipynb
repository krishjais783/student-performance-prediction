## Notebook 2: Data Preprocessing

**Objective:**
Prepare the dataset for machine learning by selecting features, splitting the data, encoding categorical variables, and scaling numerical features.

```python
import pandas as pd
df=pd.read_csv("../data/student-por.csv")
```

```python
x=df.drop("G3",axis=1)
y=df["G3"]
```

### Observation

- The target variable is G3 (Final Grade).
- All remaining columns are used as input features.

### Insight

- Separating features and the target variable is necessary before training a machine learning model.

### Possible Action

- Use X for model training and y as the prediction target.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

### Observation

- The dataset was split into training and testing sets
- 80% of the data is used for training and 20% for testing

### Insight

- Keeping a separate test set helps evaluate how well the model performs on unseen data

### Possible Action

- Use the training set to fit the model and the test set to evaluate its performance

```python
categorical_col=X_train.select_dtypes(include="object").columns.tolist()
categorical_col
```

```python
numerical_col=X_train.select_dtypes(exclude="object").columns.tolist()
numerical_col
```

### Observation

- The dataset contains both categorical and numerical features
- Categorical features include variables like sex, school, and internet
- Numerical features include variables like age, studytime, absences, G1, and G2

### Insight

- Different preprocessing techniques are required for categorical and numerical data
- Separating these columns makes it easier to build a preprocessing pipeline

### Possible Action

- Apply One-Hot Encoding to categorical features
- Apply feature scaling to numerical features

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

preprocessor = ColumnTransformer(transformers=[("cat", OneHotEncoder(handle_unknown="ignore"), categorical_col),("num", StandardScaler(), numerical_col)])

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

### Observation

- Categorical features were encoded using One-Hot Encoding
- Numerical features were scaled using StandardScaler
- The preprocessing was applied separately to the training and testing datasets

### Insight

- Encoding converts categorical values into a format suitable for machine learning models
- Scaling ensures that numerical features are on a similar scale, preventing larger values from dominating the model

### Possible Action

- Use the processed training and testing data for machine learning model training
