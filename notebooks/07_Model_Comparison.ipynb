# Student Performance Analytics & Prediction

## Notebook 7: Model Comparison

### Objective

Compare the performance of Linear Regression, Decision Tree, and Random Forest using regression metrics and identify the best-performing model for deployment.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comparison_df = pd.DataFrame({
    "Model": ["Linear Regression","Decision Tree","Random Forest"],
    "MAE": [0.77, 0.82, 0.75],
    "RMSE": [1.21, 1.58, 1.24],
    "R² Score": [0.85, 0.74, 0.84]})
    
comparison_df
```

## Performance Comparison

### Observation

- The table compares all three regression models using MAE, RMSE, and R² Score.

### Insight

- Lower MAE and RMSE indicate better prediction accuracy.
- Higher R² Score indicates a better fit to the data.
- These metrics help identify the most suitable model for deployment.

```python
plt.figure(figsize=(6,4))
plt.bar(comparison_df["Model"], comparison_df["R² Score"])

plt.title("R² Score Comparison")
plt.ylabel("R² Score")

plt.grid(axis="y")
plt.show()
```

## R² Score Analysis

### Observation

- The R² scores of the three models are compared.

### Insight

- A higher R² score indicates that the model explains more variation in students' final grades.

### Possible Action

- Choose the model with the highest R² score if overall prediction quality is the priority.

```python
plt.figure(figsize=(6,4))
plt.bar(comparison_df["Model"], comparison_df["MAE"])

plt.title("MAE Comparison")
plt.ylabel("Mean Absolute Error")

plt.grid(axis="y")
plt.show()
```

## MAE Analysis

### Observation

- The average prediction error of each model is compared.

### Insight

- Lower MAE means predictions are closer to the actual grades.

### Possible Action

- Prefer models with lower MAE for more accurate predictions.

```python
plt.figure(figsize=(6,4))
plt.bar(comparison_df["Model"], comparison_df["RMSE"])

plt.title("RMSE Comparison")
plt.ylabel("Root Mean Squared Error")

plt.grid(axis="y")
plt.show()
```

## RMSE Analysis

### Observation

- The RMSE values of all models are compared.

### Insight

- Lower RMSE indicates fewer large prediction errors.

### Possible Action

- Select models with lower RMSE for more reliable performance.

## Best Model Selection

### Observation

- Random Forest achieved the lowest MAE, while Linear Regression achieved the highest R² score. Decision Tree showed comparatively lower performance.

### Insight

- Both Linear Regression and Random Forest performed well on this dataset. The choice depends on the deployment requirements and the balance between     
  prediction accuracy and model complexity.

### Final Decision

- Random Forest achieved the lowest MAE, while Linear Regression achieved the highest R² score and slightly lower RMSE.

- Since Random Forest provides the lowest average prediction error while maintaining performance close to Linear Regression, it is selected as the   
  final model for deployment.
